```python
df = df.dropna(subset=["Description"])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df_with_customer=df.dropna(subset=["CustomerID"]).copy()
df_with_customer["CustomerID"]=df_with_customer["CustomerID"].astype(int)
print(df_with_customer)
```

```python
df_with_customer["CustomerID"]
```

```python
df["IsCancelled"] = df["InvoiceNo"].astype(str).str.startswith("C").sum()
```

```python
df["IsCancelled"]
```

```python
df_clean = df[
    (~df["IsCancelled"]) &   
    (df["Quantity"] > 0) &   
    (df["UnitPrice"] > 0)    
].copy()  
```

```python
df_clean["InvoiceDate"] = pd.to_datetime(df_clean["InvoiceDate"], format="%m/%d/%Y %H:%M")
```

```python
df_clean["InvoiceDate"]
```

```python

df_clean["TotalPrice"] = np.multiply(df_clean["Quantity"], df_clean["UnitPrice"])
```

```python
df_clean["TotalPrice"]
```

```python
print("Original rows:", df.shape[0], "| Clean rows:", df_clean.shape[0])
print("Cancelled orders removed:", df["IsCancelled"].sum())
print(df_clean[["InvoiceNo", "Quantity", "UnitPrice", "TotalPrice", "InvoiceDate"]].head())
print("\n")
```

```python
df_clean["Year"] = df_clean["InvoiceDate"].dt.year
df_clean["Month"] = df_clean["InvoiceDate"].dt.month
df_clean["Day"] = df_clean["InvoiceDate"].dt.day
df_clean["Hour"] = df_clean["InvoiceDate"].dt.hour
df_clean["DayOfWeek"] = df_clean["InvoiceDate"].dt.day_name()
```

```python
print(df_clean[["InvoiceDate", "Year", "Month", "DayOfWeek", "Hour"]].head(10))
print("\n")
```

```python
total_revenue = np.sum(df_clean["TotalPrice"])
total_orders = df_clean["InvoiceNo"].nunique()
total_customers = df_clean["CustomerID"].nunique()
avg_order_value = df_clean.groupby("InvoiceNo")["TotalPrice"].sum().mean()
```

```python
print(f"Total Revenue: Â£{total_revenue:,.2f}")
print(f"Total Unique Orders: {total_orders}")
print(f"Total Unique Customers: {total_customers}")
print(f"Average Order Value: Â£{avg_order_value:,.2f}")
print("\n")
```

```python
top_products_revenue = (
    df_clean.groupby("Description")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products_qty = (
    df_clean.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
print("Top 10 Products by Revenue")
print(top_products_revenue)
print("\nTop 10 Products by Quantity Sold")
print(top_products_qty)
print("\n")
```

```python
country_revenue = (
    df_clean.groupby("Country")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
)
```

```python
print("STEP 7: Revenue by Country (Top 10)")
print(country_revenue.head(10))
print("\n")
```

```python
df_clean["YearMonth"] = df_clean["InvoiceDate"].dt.to_period("M")
monthly_sales = df_clean.groupby("YearMonth")["TotalPrice"].sum()
```

```python
print("Monthly Revenue Trend")
print(monthly_sales)
print("\n")
```

```python
customer_summary = df_clean.groupby("CustomerID").agg(
    TotalSpent=("TotalPrice", "sum"),
    NumOrders=("InvoiceNo", "nunique"),
    AvgOrderValue=("TotalPrice", "mean"),
    LastPurchase=("InvoiceDate", "max")
).sort_values("TotalSpent", ascending=False)
```

```python
print("Top 10 Customers by Total Spend")
print(customer_summary.head(10))
print("\n")
```

```python
order_values = df_clean.groupby("InvoiceNo")["TotalPrice"].sum().values
```

```python
print("Order Value Statistics (NumPy)")
print(f"Mean: Â£{np.mean(order_values):,.2f}")
print(f"Median: Â£{np.median(order_values):,.2f}")
print(f"Std Dev: Â£{np.std(order_values):,.2f}")
print(f"Min: Â£{np.min(order_values):,.2f}")
print(f"Max: Â£{np.max(order_values):,.2f}")
print(f"25th Percentile: Â£{np.percentile(order_values, 25):,.2f}")
print(f"75th Percentile: Â£{np.percentile(order_values, 75):,.2f}")
print("\n")
```

```python
df_clean.to_csv("cleaned_ecommerce_data.csv", index=False)
customer_summary.to_csv("customer_summary.csv")
country_revenue.to_csv("country_revenue.csv")
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.figure()
monthly_sales.plot(kind='line', marker='o',color='royalblue')
plt.title('Monthly Revenue Trend',fontsize=16,fontweight='bold')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue (Â£)')
plt.xticks(rotation=45)
plt.savefig('monthly_revenue_trend.png', bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
        x=top_products_revenue.index,
        y=top_products_revenue.values,
        palette='viridis'
)
plt.title('Top 10 Products by Revenue', fontsize=14, fontweight='bold')
plt.xlabel("Total Revenue ($)")
plt.ylabel("Product") 
plt.tight_layout()
plt.savefig("top_10_products_revenue.png", dpi=150)
plt.show()
```

```python
op_countries = country_revenue.head(10)
plt.figure()
sns.barplot(
        x=op_countries.index,
       y=op_countries.values,
        hue=op_countries.index,
        palette="crest"
)
plt.title('Top 10 Countries by Revenue', fontsize=14, fontweight='bold')
plt.xlabel("Total Revenue ($)")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("top_10_countries_revenue.png", dpi=150)
plt.show()
```

```python
plt.figure()
sns.histplot(order_values, bins=50,kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Order Values', fontsize=14, fontweight='bold')
plt.xlabel("Order Value (Â£)")
plt.ylabel("Frequency")
plt.xlim(0, 1000)  # Adjust the x-axis limits if needed
plt.tight_layout()
plt.savefig("order_value_distribution.png", dpi=150)
plt.show()
```

```python
day_order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
sales_by_day = df_clean.groupby("DayOfWeek")["TotalPrice"].sum().reindex(day_order)
sns.barplot(x=sales_by_day.index,y=sales_by_day.values,hue=sales_by_day.index,palette="coolwarm")
plt.title('Total Revenue by Day of the Week', fontsize=14, fontweight='bold')
plt.xlabel("Day of the Week")
plt.ylabel("Total Revenue (Â£)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("revenue_by_day_of_week.png", dpi=150)
plt.show()
```

```python
top_customer=customer_summary.head()
plt.figure()
sns.barplot(
        x=top_customer["TotalSpent"].values,
        y=top_customer.index.astype(str),
        palette="flare"
)
plt.title('Top 10 customer by total spend',fontsize=14,fontweight="bold")
plt.xlabel("total spent")
plt.ylabel("Customer id")
plt.tight_layout()
plt.savefig("top_customer.png",dpi=150)
plt.show()
```
